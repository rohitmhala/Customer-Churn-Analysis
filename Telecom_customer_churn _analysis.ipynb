```python
def conv(value):
    if value==1:
        return "yes"
    else :
        return "No"
data["SeniorCitizen"]=data["SeniorCitizen"].apply(conv)
```

```python
ax=sns.countplot(x='Churn',data=data,color="red")                          #no. of customer churn
ax.bar_label(ax.containers[0])
plt.title("Count of Number of Customer Churn")
#counting overall churn
```

```python
gb=data.groupby("Churn").agg({"Churn":"count"})                                       # churn in percentage
gb
plt.pie(gb['Churn'],labels=['yes','NO'],autopct='%0.1f%%')
plt.title("Percentage of Customer Churn")
plt.show()
```

```python
sns.countplot(x="gender",data=data,hue="Churn")                   #no of male /female churning or not
plt.title("Gender Specific Churn")
plt.show()
```

```python
sns.countplot(x="SeniorCitizen",data=data,hue="Churn")                   #no of senior citizens churning or not
plt.title("SeniorCitizen Specific Churn")
plt.show()
```

```python
# Step 1: Create percentage table
table = pd.crosstab(data['SeniorCitizen'], data['Churn'], normalize='index') * 100            #greater percentage of data of senior citizen is churned

# Step 2: Plot stacked bar chart
table.plot(kind='bar', stacked=True, color=['skyblue', 'orange'])

# Step 3: Chart formatting
plt.title("SeniorCitizen Specific Churn (%)")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.legend(title='Churn')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
ax=sns.countplot(x="Contract",data=data,hue="Churn")                   #contract specific churn longer the contract lesser the churn
ax.bar_label(ax.containers[0])
plt.title("contract Specific Churn")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure",data=data,hue="Churn",bins=72)       #pepole who have long time tenure or contract they stay and people who use only one two month are churnout
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure",data=data,hue="Churn",bins=72,kde=True)       #pepole who have long time tenure or contract they stay and people who use only one two month are churnout showing in terms of frequency
plt.show()
```

```python
# Example: Load your DataFrame (replace this with your actual data)
# data = pd.read_csv("your_file.csv")

# List of columns to plot
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Setup subplots
plt.figure(figsize=(18, 20))
for i, col in enumerate(cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=data, x=col, palette="rocket",hue="Churn")
    plt.title(col)
    plt.xlabel('')
    plt.ylabel('Count')
    plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```

#The subplots display countplots of various service-related features split by customer churn status. It shows that customers with no additional services (like OnlineSecurity, TechSupport, etc.) are more likely to churn. Features such as StreamingTV, DeviceProtection, and OnlineBackup also show higher churn rates among non-users. This indicates that value-added services may help reduce customer churn.
